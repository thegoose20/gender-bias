```python
def getLabelNumber(df, col, label, label_dict=label_to_no):
    values = list(df[col])
    new_values = []
    for value in values:
        if label in value:
            new_values += [label_dict[label]]
        else:
            new_values += [0]
    return new_values
x = getLabelNumber(baseline_osc_preds, "manual_label", "Omission")
y = getLabelNumber(baseline_osc_preds, "manual_label", "Stereotype")
print(x[:20])
print(y[:20])
```

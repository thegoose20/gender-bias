#### Normalization 
Lemmatize the tokens:

```python
lmtzr = WordNetLemmatizer()
```

```python
tokens_train = list(df_train.token)
lemmas_train = [lmtzr.lemmatize(token) for token in tokens_train]
tokens_dev = list(df_dev.token)
lemmas_dev = [lmtzr.lemmatize(token) for token in tokens_dev]
```

```python
df_train.insert((list(df_train.columns).index("token")+1), "lemma", lemmas_train)
df_dev.insert((list(df_dev.columns).index("token")+1), "lemma", lemmas_dev)
```

#### Part-of-Speech Tagging

```python
df_sent = pd.read_csv(config.tokc_path+"sentences.csv",index_col=0)
print(df_sent.shape)
# Ignore sentences that aren't in input data
df_test = pd.read_csv(config.tokc_path+"model_input/token_test.csv", index_col=0)
df_test = df_test.loc[df_test.field != "Identifier"]
all_sentids = []
for df in [df_train, df_dev, df_test]:
    sentids = set(list(df.sentence_id.unique()))
    all_sentids += list(sentids)
all_sentids = list(set(all_sentids))
df_sent = df_sent.loc[df_sent.sentence_id.isin(all_sentids)]  
print(df_sent.shape)
df_sent.head()
```

```python
sentences = list(df_sent.sentences)
tokenized_sentences = [nltk.word_tokenize(s) for s in sentences]
tagged_sentences = [nltk.pos_tag(s) for s in tokenized_sentences]
print(sentences[0])
print(tokenized_sentences[0])
print(tagged_sentences[0])
```
